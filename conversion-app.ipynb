```python
import numpy as np  # Importing numpy library
import pandas as pd  # Importing pandas library

import os
# os.walk method is used to traverse directory tree in a top-down approach
# it returns the root directory, directories in the root and files in each directory
# Here, we are using it to find files in the "/kaggle/input" directory
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))  # Print file path
```

```python
n = 0  # Initialize n with value 0
print(n)  # Prints the value of n

n = n + 4  # Adds 4 to n, new value of n is 4
if n > 0:  # Check if n is greater than 0
    print("Vincida")  # If yes, print "Vincida"

x = "Bangladeshi " * n  # Repeat "ruhingya" n times and assign it to x
print(x)  # Print the value of x
```

```python
spam_amount = 0  # Initialize spam_amount with value 0
print(spam_amount)  # Prints the value of spam_amount

spam_amount = spam_amount + 6  # Add 6 to spam_amount, new value of spam_amount is 6
if spam_amount > 0:  # Check if spam_amount is greater than 0
    print("But I don't want ANY spam!")  # If yes, print "But I don't want ANY spam!"
```

```python
hat_height_cm = 25  # Initialize hat_height_cm with value 25
my_height_cm = 190  # Initialize my_height_cm with value 190
total_height_meters = (hat_height_cm + my_height_cm) / 100  # Calculate total height in meters, add hat height and my height and divide by 100
print("My height is:", total_height_meters, "meters")  # Print the value of total_height_meters with annotation "My height is:"
```

```python
# Basic Data Types Challenge 2: MPH to MPS Conversion App

print("Welcome to the MPH to MPS Conversion App")

# Gather user input
mph = float(input("\nWhat is your speed in miles per hour: "))

# Convert to meters per second (mps)
mps = mph * 0.4474

# Display the result to the user
print("Your speed in meters per second is " + str(mps) + ".")
```
